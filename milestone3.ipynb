```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM, Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                           roc_auc_score, roc_curve, confusion_matrix,
                           classification_report, precision_recall_curve, auc)
import warnings
warnings.filterwarnings('ignore')
```

```python
print("MILESTONE 3: COMPREHENSIVE MODEL EVALUATION AND BENCHMARKING")
```

```python
print("STEP 1: LOADING AND PREPARING DATA")
```

```python
df = pd.read_csv('hateXplain.csv')
```

```python
print(f"Dataset loaded: {df.shape[0]} records, {df.shape[1]} features")
```

```python
df_clean = df.copy()
df_clean.head()
```

```python
df_clean['target'].fillna('Unknown',inplace=True)
df_clean = df_clean.dropna(subset=['post_tokens'])
df_clean.head()
```

```python
df_clean = df_clean.drop_duplicates()
df_clean.head()
```

```python
df_clean = df_clean.reset_index(drop=True)
df_clean.head()
```

```python
print(f"After cleaning: {df_clean.shape[0]} records")
```

```python
le_label = LabelEncoder()
df_clean['label_encoded'] = le_label.fit_transform(df_clean['label'])
```

```python
print(f"\nLabel Distribution:")
print(df_clean['label'].value_counts())
print(f"\n Label Encoding:{dict(enumerate(le_label.classes_))}")
```

```python
    print("STEP 2: TEXT FEATURE EXTRACTION")
```

```python
tfidf_vectorizer = TfidfVectorizer(
    max_features=1000,
    min_df = 2,
    max_df = 0.8,
    ngram_range = (1,2),
    stop_words = 'english'
)
```

```python
X_tfidf = tfidf_vectorizer.fit_transform(df_clean['post_tokens'])
X_tfidf_dense = X_tfidf.toarray()
y = df_clean['label_encoded']
```

```python
print(f"TF-IDF Feature Matrix Shape: {X_tfidf_dense.shape}")
print(f"Number of features: {X_tfidf_dense.shape[1]}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_tfidf_dense, y, test_size=0.2, random_state=42, stratify=y)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
print(f"Training set size: {X_train.shape[0]}")
```

```python
print(f"Testing set size: {X_test.shape[0]}")
```

```python
print("STEP 3: EVALUATING CLASSICAL MODELS")
```

```python
classical_models = {
    'KNN': KNeighborsClassifier(n_neighbors=5),
    'Decision Tree': DecisionTreeClassifier(random_state=42, max_depth=20),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42, max_depth=20),
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'SVM': SVC(kernel='rbf', random_state=42, probability=True, max_iter=5000),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, random_state=42)
}
```

```python
classical_results = {}
```

```python
for model_name, model in classical_models.items():
    print(f"Training {model_name}...")

    if model_name in ['KNN', 'Logistic Regression', 'SVM']:
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
        y_proba = model.predict_proba(X_test_scaled) if hasattr(model, 'predict_proba') else None
    else:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        y_proba = model.predict_proba(X_test) if hasattr(model, 'predict_proba') else None
    
    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)
    
    # ROC AUC calculation
    try:
        if y_proba is not None and y_proba.shape[1] > 1:
            roc_auc = roc_auc_score(y_test, y_proba, average='weighted', multi_class='ovr')
        else:
            roc_auc = None
    except:
        roc_auc = None
    
    classical_results[model_name] = {
        'model': model,
        'y_pred': y_pred,
        'y_proba': y_proba,
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'roc_auc': roc_auc
    }
    
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
    if roc_auc is not None:
        print(f"ROC AUC: {roc_auc:.4f}")
    
    print(f"\nClassification Report:")
    print(classification_report(y_test, y_pred, target_names=le_label.classes_))
```

```python
print("STEP 4: EVALUATING DEEP LEARNING MODELS")
```

```python
deep_learning_results = {}
```

```python
print("Training Deep Learning Model 1: Multi-Layer Perceptron (MLP)")
```

```python
model_mlp = Sequential([
    Dense(256, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    Dropout(0.3),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(len(le_label.classes_), activation='softmax')
])
```

```python
model_mlp.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
print("MLP Model Architecture:")
model_mlp.summary()
```

```python
history_mlp = model_mlp.fit(X_train_scaled, y_train, epochs=30, batch_size=32, validation_split=0.2, verbose=0)
```

```python
y_pred_mlp = model_mlp.predict(X_test_scaled)
y_pred_mlp = np.argmax(y_pred_mlp, axis=1)
y_proba_mlp = model_mlp.predict(X_test_scaled)
```

```python
accuracy_mlp = accuracy_score(y_test, y_pred_mlp)
precision_mlp = precision_score(y_test, y_pred_mlp, average='weighted', zero_division=0)
recall_mlp = recall_score(y_test, y_pred_mlp, average='weighted', zero_division=0)
f1_mlp = f1_score(y_test, y_pred_mlp, average='weighted', zero_division=0)
```

```python
try:
  roc_auc_mlp = roc_auc_score(y_test, y_proba_mlp, average='weighted', multi_class='ovr')
except:
  roc_auc_mlp = None
```

```python
deep_learning_results['MLP'] = {
    'model': model_mlp,
    'y_pred': y_pred_mlp,
    'y_proba': y_proba_mlp,
    'accuracy': accuracy_mlp,
    'precision': precision_mlp,
    'recall': recall_mlp,
    'f1': f1_mlp,
    'roc_auc': roc_auc_mlp,
    'history': history_mlp
}
```

```python
print(f"MLP - Accuracy: {accuracy_mlp:.4f}, Precision: {precision_mlp:.4f}, Recall: {recall_mlp:.4f}, F1: {f1_mlp:.4f}")
if roc_auc_mlp is not None:
    print(f"ROC AUC: {roc_auc_mlp:.4f}")
```

```python
print("Classification Report for MLP:")
print(classification_report(y_test, y_pred_mlp, target_names=le_label.classes_))
```

```python
model_cnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
history_cnn = model_cnn.fit(X_train_scaled, y_train, epochs=30, batch_size=32, validation_split=0.2, verbose=0)
```

```python
y_pred_cnn = model_cnn.predict(X_test_scaled)
y_pred_cnn = np.argmax(y_pred_cnn, axis=1)
y_proba_cnn = model_cnn.predict(X_test_scaled)
```

```python
try:
  roc_auc_cnn = roc_auc_score(y_test, y_proba_cnn, average='weighted', multi_class='ovr')
except:
  roc_auc_cnn = None
```

```python
deep_learning_results['Dense Network (L2)']={
    'model': model_cnn,
    'y_pred': y_pred_cnn,
    'y_proba': y_proba_cnn,
    'accuracy': accuracy_cnn,
    'precision': precision_cnn,
    'recall': recall_cnn,
    'f1': f1_cnn,
    'roc_auc': roc_auc_cnn,
    'history': history_cnn
}
```

```python
print(f"Dense Network (L2) - Accuracy: {accuracy_cnn:.4f}, Precision: {precision_cnn:.4f}, Recall: {recall_cnn:.4f}, F1: {f1_cnn:.4f}")
if roc_auc_cnn is not None:
    print(f"ROC AUC: {roc_auc_cnn:.4f}")
```

```python
fig , axes = plt.subplots(2,3,figsize=(18,10))
axes = axes.flatten()
plot_idx = 0
for model_name , results in all_results.items():
  if results['y_proba'] is not None:
    if plot_idx >= len(axes):
            break
    ax = axes[plot_idx]
    y_test_oneshot = np.eye(len(le_label.classes_))[y_test]
    y_proba_reshaped = results['y_proba']
    
    for i, class_label in enumerate(le_label.classes_):
      fpr, tpr,_ = roc_curve(y_test_oneshot[:,i], y_proba_reshaped[:,i])
      roc_auc_i = auc(fpr,tpr)
      ax.plot(fpr, tpr, label=f'{class_label} (AUC = {roc_auc_i:.3f})')

    ax.plot([0,1],[0,1],'k--', label = 'Random', linewidth=1)
    ax.set_xlim([0.0,1.0])
    ax.set_ylim([0.0,1.05])
    ax.set_xlabel('False Positive Rate',fontweight='bold')
    ax.set_ylabel('True Positive Rate',fontweight='bold')
    ax.set_title(f'ROC Curve for {model_name}',fontweight='bold',fontsize=16)
    ax.legend(loc='lower right',fontsize = 8)
    ax.grid (alpha=0.3)


    plot_idx += 1
for idx in range(plot_idx, len(axes)):
  fig.delaxes(axes[idx]) 
plt.tight_layout()
plt.savefig('roc_curves_all_models.png',dpi=300, bbox_inches='tight')
plt.show()
```
